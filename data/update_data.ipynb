```python
#------------------------------------------------------------------------------------
# Population is monthly, we make it quarterly by averaging
#------------------------------------------------------------------------------------
data_pop = quandl.get('FRED/POP', authtoken=auth_code)
data_pop.columns = ['pop']
date_index = date_range_quarter(data_pop.index)     # see the userfunction above
data_pop = data_pop.groupby([data_pop.index.year, data_pop.index.quarter]).mean()  
data_pop.index = date_index

#------------------------------------------------------------------------------------
# Personal consumption expenditures with prices: Nondurable + Services
#------------------------------------------------------------------------------------
dataset = quandl.get(['FRED/PCND',                  # consumption exp: non-durables (quant index)
                      'FRED/PCESV',                 # consumption exp: services (quant index)
                      'FRED/DNDGRA3Q086SBEA',       # nominal consumption exp on non-durables
                      'FRED/DSERRA3Q086SBEA',       # nominal consumption exp on services
                      'FRED/DNDGRG3Q086SBEA',       # price index of consumption exp: non-durables
                      'FRED/DSERRG3Q086SBEA'],      # price index of consumption exp: services
                      authtoken=auth_code)

dataset.columns = ['consnond', 'consserv', 'qnond', 'qserv', 'pinond', 'piserv']
dataset.index = date_range_quarter(dataset.index)   # see the userfunction above

dataset = pd.concat([data_pop, dataset], axis =1)
dataset.to_csv(path_fromweb + "dataset.csv", sep = ",")
```

**(2) Merge the McCullock and Kwon with the GSW dataset **

Download the McCullock and Kwon data

```python
#================================================================================
# Download the McCulloch and Kwon file from web (and save it as txt)
#================================================================================

mcck_url = 'http://www.econ.ohio-state.edu/jhm/ts/mcckwon/zeroyld1'
ff = urllib.request.urlopen(mcck_url).read() 

text_file = open(path_fromweb + "McCullochKwon.txt", "w")
text_file.write(ff.decode('utf8'))
text_file.close()

with open(path_fromweb + "McCullochKwon.txt", 'r+') as f:
        mcculloch_data = f.readlines()    # create a list with the lines

#-------------------------------------------------------------
# Clean up the mess (create pandas dataframe called mcculloch)
#-------------------------------------------------------------

numb_lines = int(len(mcculloch_data)/8)
store_data = np.zeros((numb_lines - 1, 62))       # to store the final data

# Clean the lines from junk characters and split into numbers
for n in range(1, numb_lines):
    xx = str(mcculloch_data[n*8:(n+1)*8])
    xx = xx.replace("\\n', '", "").replace('\\n', "").replace("['", '').replace("']", '')
    for col, var in enumerate(xx.rsplit(   )):
        store_data[n - 1, col] = float(var)

# Define the DataFrame with the first line being the header
mcculloch = pd.DataFrame(data = store_data, 
                         index = pd.date_range('1946-12-01', '1985-09-01', freq = 'M'))
xx = str(mcculloch_data[:8])
xx = xx.replace("\\n', '", "").replace('\\n', "").replace("['", '').replace("']", '')
mcculloch.columns = xx.rsplit()
mcculloch = mcculloch.replace(-88.888, np.nan)
```

Download the Gurkaynak-Sack-Wright (GSW) data

```python
#------------------------------------------------------------------------------
# GSW dataset (zero-coupon, par yield, TIPS zero-coupon, TIPS par yield): 
#------------------------------------------------------------------------------
gsw_zerocoupon = quandl.get('FED/SVENY', authtoken=auth_code)
gsw_paryield = quandl.get('FED/SVENPY', authtoken=auth_code)
gsw_tipsy = quandl.get('FED/TIPSY', authtoken=auth_code)

date_index = date_range_month(gsw_zerocoupon.index)
gsw_zerocoupon = gsw_zerocoupon.groupby([gsw_zerocoupon.index.year, gsw_zerocoupon.index.month]).mean()
gsw_zerocoupon.index = date_index

date_index = date_range_month(gsw_paryield.index)
gsw_paryield = gsw_paryield.groupby([gsw_paryield.index.year, gsw_paryield.index.month]).mean()
gsw_paryield.index = date_index

date_index = date_range_month(gsw_tipsy.index)
gsw_tipsy = gsw_tipsy.groupby([gsw_tipsy.index.year, gsw_tipsy.index.month]).mean()
gsw_tipsy.index = date_index


#========================================
# Merge the McCulloch and GSW datasets
#========================================
T_month = len(pd.date_range('1952-06-30', gsw_zerocoupon[-1:].index[0], freq = 'M'))
merged_yields = np.zeros((T_month, 15))

for y in range(1, 8):
    gsw_label = 'SVENY0{:d}'.format(y)
    mcc_label = '{:d}.000'.format(y)
    merged_yields[:, y-1] = np.hstack([mcculloch[mcc_label]['1952-06-30':'1961-05-31'].values, 
                                       gsw_zerocoupon[gsw_label]['1961-06-01':].values])
for y in range(8, 16):
    if y < 10:
        gsw_label = 'SVENY0{:d}'.format(y)
    else:
        gsw_label = 'SVENY{:d}'.format(y)
    mcc_label = '{:d}.000'.format(y)
    merged_yields[:, y-1] = np.hstack([mcculloch[mcc_label]['1952-06-30':'1971-10-31'].values, 
                                       gsw_zerocoupon[gsw_label]['1971-11-01':].values])

gsw_plot = ['SVENY0{:d}'.format(i) for i in range(1, 10)] + ['SVENY{:d}'.format(i) for i in range(10, 16)]
gsw_zerocoupon = pd.DataFrame(data = merged_yields, columns=gsw_plot, 
                              index = pd.date_range('1952-06-30', gsw_zerocoupon[-1:].index[0], freq = 'M'))

gsw_tipsy_adj = pd.concat([pd.read_csv(path_data + 'old/tipsy_early.csv', sep = ',', 
                                       parse_dates = False, index_col = 0), 
                           gsw_tipsy.loc['2004':, 'TIPSY05':'TIPSY20']])
gsw_tipsy_adj.index = pd.date_range(gsw_tipsy.index[0], gsw_tipsy.index[-1], freq='M')

data_gsw = pd.concat([gsw_zerocoupon, gsw_paryield, gsw_tipsy_adj], axis = 1)
```

**(3) Treasury par (CMT) yield data to compare it with survey expectations (and for "bootstrapping")**

Because the survey questions refer to the treasury par (CMT) yields, when computing statistics involving both survey expectations and realized series, it is important to use the yield data from the Treasury.

Because the surveys take place always in the last few days of the quarter, we don't want to use quarterly averages when computing expected changes. For this reason, we include two types of quarterly treasurey par yield series 
 * (1) quarterly by averaging
 * (2) quarterly by taking the last day (append `_LD` to the respective variable name) 

```python
#-------------------------------------------------
# Treasury (CMT or par) nominal yields: 
#-------------------------------------------------
data_treasury = quandl.get(['FED/RIFLGFCM01_N_B', 'FED/RIFLGFCM03_N_B', 'FED/RIFLGFCM06_N_B',
                            'FED/RIFLGFCY01_N_B', 'FED/RIFLGFCY02_N_B', 'FED/RIFLGFCY03_N_B', 
                            'FED/RIFLGFCY05_N_B', 'FED/RIFLGFCY07_N_B', 'FED/RIFLGFCY10_N_B', 
                            'FED/RIFLGFCY20_N_B', 'FED/RIFLGFCY30_N_B'], authtoken=auth_code)

data_treasury.columns = ['TRNPM01', 'TRNPM03', 'TRNPM06', 'TRNPY01', 'TRNPY02', 'TRNPY03', 
                         'TRNPY05', 'TRNPY07', 'TRNPY10', 'TRNPY20', 'TRNPY30']

# First create quarterly series by taking the last obervation each quarter
last_days = data_treasury.groupby([data_treasury.index.year, data_treasury.index.quarter]).last()
last_days.index = date_range_quarter(data_treasury.index)
last_days.columns = ['TRNPM01_LD','TRNPM03_LD','TRNPM06_LD','TRNPY01_LD','TRNPY02_LD','TRNPY03_LD',
                     'TRNPY05_LD','TRNPY07_LD','TRNPY10_LD','TRNPY20_LD','TRNPY30_LD']

date_index = date_range_month(data_treasury.index)
data_treasury = data_treasury.groupby([data_treasury.index.year, data_treasury.index.month]).mean()
data_treasury.index = date_index
data_treasury.index = pd.date_range(data_treasury.index[0], data_treasury.index[-1], freq='M')

#-------------------------------------------------
# Treasury (CMT or par) real yields: 
#-------------------------------------------------
treasury_real = quandl.get('USTREASURY/REALYIELD', authtoken=auth_code)
treasury_real.columns = ['TRRPY05', 'TRRPY07', 'TRRPY10', 'TRRPY20', 'TRRPY30']
date_index = date_range_month(treasury_real.index)
treasury_real = treasury_real.groupby([treasury_real.index.year, treasury_real.index.month]).mean()
treasury_real.index = date_index

yields_monthly = pd.concat([data_gsw, data_treasury, treasury_real], axis = 1)
yields_monthly.to_csv(path_fromweb + 'yields_monthly.csv', sep = ',')
```

## Computing the consumption growth and inflation rate series

Note:
  * $1952: Q2 - \text{present}$ -- when the CRSP yield data starts
  * $1961: Q2 - \text{present}$ -- when the GSW yield data starts
  * $1982: Q2 - \text{present}$ -- when the Blue-Chip survey data starts
  * $1952: Q2 - 2005:Q4$ -- Piazzesi and Schneider's sample period

```python
dataset = pd.read_csv(path_fromweb + 'dataset.csv', sep = ',', parse_dates = False, index_col = 0 )
dataset = dataset[np.isfinite(dataset['pop']) & np.isfinite(dataset['qnond'])]

# Weigths for the nondurable consumption and services
wndurable = (dataset['consnond'] / (dataset['qnond'] * dataset['pinond'])).mean()
wservice = (dataset['consserv'] / (dataset['qserv'] * dataset['piserv'])).mean()
w = np.asarray([[wndurable, wservice]])
PI = np.asarray(dataset[['pinond', 'piserv']])
QI = np.asarray(dataset[['qnond', 'qserv']])
POP = np.asarray(dataset['pop'])

#----------------------------------------------------------------------------
# Formulas for the price/consumption growth rates come from the NIPA manual 

p1q1 = (w * PI[1:, :] * QI[1:, :]).sum(1)
p0q0 = (w * PI[:-1, :] * QI[:-1, :]).sum(1)
p1q0 = (w * PI[1:, :] * QI[:-1, :]).sum(1)
p0q1 = (w * PI[:-1, :] * QI[1:, :]).sum(1)

p_gr = np.hstack([1, np.sqrt((p1q0 * p1q1)/(p0q0 * p0q1))])
c_gr = np.hstack([1, np.sqrt((p0q1 * p1q1)/(p0q0 * p1q0))])
pop_gr = np.hstack([1, POP[1:]/POP[:-1]])

#----------------------------------------------------------------------------
# Indices from which growth rates will be calculated

p_ind = np.cumprod(p_gr)
c_ind = np.cumprod(c_gr)
pop_ind = np.cumprod(pop_gr)
#pop_ind = np.ones(len(c_ind))

log_p_ind = np.log(p_ind)
log_c_ind = np.log(c_ind / pop_ind)

dc = 100*(log_c_ind[1:] - log_c_ind[:-1])
pi = 100*(log_p_ind[1:] - log_p_ind[:-1])

#----------------------------------------------------------------------------

d = {'dc': dc, 'pi': pi}
data_cons = pd.DataFrame(data = d, index = pd.date_range('1952-06-30', periods = dc.shape[0], freq = 'Q'))
```

### Calculating quarterly yields

Three sources of zero coupon yields: 

    * CRSP: risk-free rate + from year1 to year 5 (from 1952:Q2)
    * McCK: McCullock and Kwon data (from 1946:Q4-1986:Q4)
    * GSW: - from year1 to year7 (from 1961:Q2 - present) 
           - from year8 to year15 (from 1971:Q3 - present) 
           - from year16 to year20 (from 1981:Q3 - present)
           - from year21 to year30 (from 1985:Q4 - present)

Two sources of par (CMT) yields: 
    * USTR:  (from 1962:Q1)
    * GSW: - from year1 to year7 (from 1961:Q2 - present) 
           - from year8 to year15 (from 1971:Q3 - present) 
           - from year16 to year20 (from 1981:Q3 - present)
           - from year21 to year30 (from 1985:Q4 - present)

Monthly data sets contain yields on the *last trading days* of the given months in CRSP and *monthly averages* in case of the other datasets. Turning the monthly data into quarterly in two different ways:
 1. Average of the three months
 2. Last month of the quarter

```python
#-----------------------------------------
# CRSP zero-coupon yields: 
#-----------------------------------------
data_crsp = pd.read_csv(path_crsp + "nom_crsp.csv", sep = ",", parse_dates = ['qdate'], index_col = 'qdate')
data_crsp.index = pd.date_range('1952-06-30', '2016-12-31', freq='M')

#-----------------------------------------
# Loading GSW and US Treasury dataset: 
#-----------------------------------------
yields_monthly = pd.read_csv(path_fromweb + 'yields_monthly.csv', sep = ',', parse_dates = True, index_col = 0 )

yields_monthly = pd.concat([data_crsp, yields_monthly], axis = 1)

#==================================================================================
# TURN EVERYTHING INTO QUARTERLY DATA 
#==================================================================================

# First method
yields_quarterly_v1 = yields_monthly.groupby([yields_monthly.index.year, yields_monthly.index.quarter]).mean()
yields_quarterly_v1.index = date_range_quarter(yields_monthly.index)

# Second method
yields_quarterly_v2 = yields_monthly.groupby([yields_monthly.index.year, yields_monthly.index.quarter]).last()
yields_quarterly_v2.index = date_range_quarter(yields_monthly.index)

data = pd.concat([data_cons, yields_quarterly_v1, last_days], axis = 1)
```

### Converting treasury par yields into zero-coupon rates ("Bootstrapping")

```python
#-------------------------------------------------------------
# Define the arguments for the FB_bootstrap function
#-------------------------------------------------------------
treas_par = ['TRNPM06','TRNPY01','TRNPY02','TRNPY03',
             'TRNPY05','TRNPY07','TRNPY10','TRNPY20','TRNPY30']

treas_par_LD = ['TRNPM06_LD','TRNPY01_LD','TRNPY02_LD','TRNPY03_LD',
                'TRNPY05_LD','TRNPY07_LD','TRNPY10_LD','TRNPY20_LD','TRNPY30_LD']

maturities_quarters = np.asarray([2, 4, 8, 12, 20, 28, 40, 80, 120])

#--------------------------------------------------------------------
# Run the bootstrap for the average and last day treasury par yields
#--------------------------------------------------------------------

zeros, zeros_LD = [], []
for t in data['1982':].index:
    zeros.append(FB_bootstrap(data[treas_par].loc[t], maturities_quarters))
    zeros_LD.append(FB_bootstrap(data[treas_par_LD].loc[t], maturities_quarters))

treasury_zeros = pd.DataFrame(data=np.hstack((np.asarray(zeros), np.asarray(zeros_LD))), 
                              index=data['1982':].index)
treasury_zeros.columns = [str(i)+'y' for i in np.arange(1, 61)/2] + [str(i)+'y_LD' for i in np.arange(1, 61)/2]
```

### Merging everything (except survey data) together

```python
data = pd.concat([data, treasury_zeros], axis = 1)
data.to_csv(path_data + "data.csv", sep = ",")
```
