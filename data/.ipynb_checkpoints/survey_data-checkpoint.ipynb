This notebook imports the survey data from various sources (csv's for the early periods, excel files for the last two decades but in different formats). For information about the availability of the different series see this [google sheet](https://docs.google.com/spreadsheets/d/1K-H-HPeUYXfzqj4_W3L06BjBvUdeo0jCQof7p6W76mg/edit#gid=0) or the `varinfo_table.csv` file

> The original BCFF data are stored in two subfolder 
>  * `csv_files` (1983-2000): naming convention `year_month_horizon.csv`, where `year` and `month` refer to the survey date
>  * `excel_files` (2000-present): there are two formats
>     - `MonthYear` + `bcffs`: period 2001-2007
>     - `BCFF` + `MonthYear` + `Excel`: period 2007-present

-------------
> Output files stored in `data/survey_data/` folder:
>  * `survey_expectations.csv`: Three dimensional dataframe of survey expectations for various assets, maturities and forecast horizons. Naming convention for the variables is: `maturity_forecasthorizon`

> Output files stored in `data/` folder:
>  * `survey_data.csv`:

```python
varname_table = pd.read_csv(path_survey + 'varinfo_table.csv', sep=',')

# Infer last date from the varinfo table
y, q = varname_table[['year', 'quarter']].iloc[-1, :]  
d = 30 if (q == [2, 3]).any() else 31
last_date = pd.datetime(y, 3*q, d)

varname_table.index = pd.date_range('1983-03-31', last_date, freq='Q')
varname_table = varname_table.drop(['year', 'quarter'], axis=1)
```

```python
def import_csv(date, horizon, data):
    year = date.year
    quarter = date.quarter

    # See the naming conventions above    
    paths = [path_survey + i for i in ['csv_files/Q1/', 
                                       'csv_files/Q2/', 
                                       'csv_files/Q3/', 
                                       'csv_files/Q4/']]
    mid = ['_01_', '_04_', '_07_','_10_']
    
    # Import the csv file
    try:
        filename = str(year) + mid[quarter-1] + str(horizon) + ".csv"
        epdata = pd.read_csv(paths[quarter-1] + filename, sep = ',', header = None, index_col = 0)
        epdata.columns = list(varname_table.loc[date].dropna())
        epdata.reset_index(drop=True, inplace=True)

        epdata['date'] = date-1           # the observation is assigned to the previous quarter
        epdata['horizon'] = horizon
        epdata.set_index(['date', 'horizon'], inplace=True)

        data = data.append(epdata)
    
        return data
    
    except FileNotFoundError:
        # If horizon does not exist, then gives back the original data DataFrame
        return data

def import_excel(date, horizon, data, filename):
    
    cols = 'A,B,D,F,H,J,L,N,P,R,T,V,X,Z,AB,AD'
    
    try:
        # Import excel sheet and make the same format as the csv files
        epdata = pd.read_excel(filename, sheetname = horizon-1, parse_cols = cols,
                               skiprows = 12, header = None, index_col = 0)
        first_row = np.where(pd.isnull(epdata.index) == True)[0][0]
        epdata = epdata.drop(epdata.index[first_row:])
        epdata = epdata.replace('na', np.nan)
        
        epdata.columns = list(varname_table.loc[date].dropna())
        epdata.reset_index(drop=True, inplace=True)
        epdata['date'] = date-1           # the observation is assigned to the previous quarter
        epdata['horizon'] = horizon
        epdata.set_index(['date', 'horizon'], inplace=True)

        data = data.append(epdata)
    
        return data
    
    except FileNotFoundError:
        # If horizon does not exist, then gives back the original data DataFrame
        return data
```

## Import survey expectations

We want to create a pandas DataFrame with the following structure
 * Multi-index row: (date, horizon)
 * Columns: asset named as in the varinfo_table.csv file

### Period 1:  csv files from 1983q1 to 2000q4



```python
#-------------------------------------------------
# Define an initial empty dataframe
#-------------------------------------------------
survey_expectations = pd.DataFrame(data=[], 
                                   columns=list(varname_table.iloc[0].dropna())+['date', 'horizon'])
survey_expectations.set_index(['date', 'horizon'], inplace=True)

#-------------------------------------------------
# Load csv files
#-------------------------------------------------
for date in varname_table.loc[:'2000'].index:
    for horizon in range(1, 7):
        survey_expectations = import_csv(date, horizon, survey_expectations)
```

### Period 2: excel files for 2000-present

Reading excel files is quite slow...

```python
path_excel = path_survey + 'excel_files/'
quart = ['01', '04','07','10'] 

for date in varname_table.loc['2001':].index:
    for horizon in range(1, 6):
        
        # Take care of the change in naming conditions
        if date.year < 2008:
            filename = path_excel + quart[date.quarter-1] + '0' + str(date.year % 2000) + 'bcffs.xls'
        else:
            filename = path_excel + 'BCFF' + quart[date.quarter-1] + str(date.year)[-2:] + 'Excel.xls'

        survey_expectations = import_excel(date, horizon, survey_expectations, filename)
```

### Compute various statistics (median, stdev, quantiles)

```python
dummy = survey_expectations.reset_index()
dummy.to_csv(path_survey + 'survey_expectations.csv', sep = ',')

median = dummy.groupby(['date', 'horizon']).median()
mean = dummy.groupby(['date', 'horizon']).mean()
stdev = dummy.groupby(['date', 'horizon']).std()
p25 = dummy.groupby(['date', 'horizon']).quantile(.25)
p75 = dummy.groupby(['date', 'horizon']).quantile(.75)
interquart = p75-p25
```

```python
# Example plot: median k-quarter-ahead forecast for the short rate 
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
for k in range(1, 6):
    median.loc[(slice('1983', '2017'), k), 'tbill_3m'].reset_index('horizon', drop=True).plot(ax=ax[0], label=str(k)+'q ahead')
    stdev.loc[(slice('1983', '2017'), k), 'tbill_3m'].reset_index('horizon', drop=True).plot(ax=ax[1], label=str(k)+'q ahead')
ax[0].set_title('Median of expectations (short rate)')
ax[1].set_title('Stdev of expectations (short rate)')
NBER_Shade(ax[0], '1983', '../data/from_web/NBER_dates.txt')
NBER_Shade(ax[1], '1983', '../data/from_web/NBER_dates.txt')

plt.legend()
```

### Save the file median forecasts

Construct the `survey_expectations.csv` dataset for the median forecasts using the following naming convention:
    
    `maturity_forecasthorizon`
    
**Notes:** 
 * First, I take a subset of the data for the forecasts about the treasury par yields 
 * I merge the `tbond_20y` and `tr_longterm_ave` into `tbond_30y`
 * Shift the dates index one quarter backward (to obtain expectatations made at the last day of the
 index quarter)


```python
# Subset of treasury par yield variables
treas_par_long = ['tbill_3m', 'tbill_6m', 'tbill_1y', 'tnote_2y', 'tnote_3y', 'tnote_5y', 
             'tnote_7y', 'tnote_10y', 'tnote_20y', 'tr_longterm_ave', 'tbond_30y']

# Merge the long-term assets and rename as tnote_30y
survey_exp = pd.concat([median[treas_par_long], 
                        median[['tnote_20y', 'tr_longterm_ave', 'tbond_30y']].mean(axis=1)], axis=1)
survey_exp.columns = treas_par_long + ['tnote_30y']
survey_exp.drop(['tnote_20y', 'tr_longterm_ave', 'tbond_30y'], axis=1, inplace=True)

dummy2 = survey_exp.reset_index()
median = dummy2.groupby(['date', 'horizon']).median()
```

```python
#-------------------------------------------------------------
# Define the arguments for the FB_bootstrap function
#-------------------------------------------------------------
treas_par = list(survey_exp.columns)
maturities_quarters = np.asarray([1, 2, 4, 8, 12, 20, 28, 40, 120])

#--------------------------------------------------------------------
# Run the bootstrap for the median treasury par yield forecasts
#--------------------------------------------------------------------
survey_treas_zeros = pd.DataFrame(index = varname_table.index-1)

for horizon in np.arange(1, 6):
    
    exp_zeros = []
    for t in varname_table.index-1:
        try:
            exp_zeros.append(FB_bootstrap(median[treas_par].loc[(t, horizon), :], maturities_quarters))
        except KeyError:
            exp_zeros.append([np.nan]*60)
    df = pd.DataFrame(data = np.asarray(exp_zeros), index = varname_table.index-1)
    df.columns = [str(i) + 'y_' + str(horizon) + 'q' for i in np.arange(1, 61)/2]
    survey_treas_zeros = pd.concat([survey_treas_zeros, df], axis = 1)
    
survey_treas_zeros.to_csv(path_data + 'data_survey_zeros.csv', sep = ',')

survey_treas_pars = dummy2.groupby(['date', 'horizon']).median()    
survey_treas_pars.to_csv(path_data + 'data_survey_pars.csv', sep = ',')
```

```python
date_slider = FloatSlider(min = 0, max = len(realized_treas_pars.index)-2, step = 1, value = 100)

@interact(date_ind = date_slider)
def expected_treasury_pars(date_ind):
    
    date = realized_treas_pars.index[int(date_ind)]

    mat_realized = [1, 2, 4, 8, 12, 20, 28, 40, 80, 120] 
    mat_survey = [1, 2, 4, 8, 12, 20, 28, 40, 120]

    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    ax[0].plot(mat_realized, nan_interpolator(realized_treas_pars.loc[date, :].T), 
           label='realized (around time of survey)', color='k', lw=3)
    ax[0].plot(mat_survey, nan_interpolator(survey_treas_pars.loc[(date, 1), :].T), 
           label='1q ahead', color='.4', lw=2)
    ax[0].plot(mat_survey, nan_interpolator(survey_treas_pars.loc[(date, 3), :].T), 
           label='3q ahead', color='.7', lw=2)
    ax[0].plot(mat_survey, nan_interpolator(survey_treas_pars.loc[(date, 5), :].T), 
           label='5q ahead', color='.9', lw=2)
    ax[0].set_title(date)
    ax[0].legend(loc='best')
    ax[0].set_ylim([0, 10])

    ax[1].plot(mat_realized, nan_interpolator(realized_treas_pars.loc[date, :].T), 
           label='realized (around time of survey)', color='k', lw=3)
    ax[1].plot(mat_survey, nan_interpolator(survey_treas_pars.loc[(date, 1), :].T), 
           label='1q ahead', color='.4', lw=2)
    ax[1].plot(mat_survey, nan_interpolator(survey_treas_pars.loc[(date, 3), :].T), 
           label='3q ahead', color='.7', lw=2)
    ax[1].plot(mat_survey, nan_interpolator(survey_treas_pars.loc[(date, 5), :].T), 
           label='5q ahead', color='.9', lw=2)
    ax[1].set_title(date)
    
    plt.tight_layout()
```
