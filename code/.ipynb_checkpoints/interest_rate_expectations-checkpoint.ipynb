```python
%matplotlib inline

import numpy as np
import scipy as sp
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

import statsmodels.api as sm
import statsmodels.tsa.api as tsa

from robust import *
from robust_utils import *

sb.set_style('white')
sb.set_style("ticks")


##=================================================
## Generate latex table
##=================================================

def format_table(x):
    if type(x) == float:
        return '{0:.3f}'.format(x)
    if type(x) == str:
        return '{0:.5s}'.format(x)

def format_table_par(x):
    if type(x) == float:
        return '({0:.3f})'.format(x)
    if type(x) == str:
        return '{0:.5s}'.format(x)
```

## Data analysis - comparison of moments 

### 1. Conditional level (data)

```python
cond_level_table = np.zeros((8, numb_bcff_vars))

for ii, var in enumerate(bcff_vars):
    nw_se = NW_se(data_survey[var], 4)
    cond_level_table[:4, ii] = nw_se[0][0], nw_se[0][1], nw_se[1][0], nw_se[1][1]

print('Conditional level (table)')
print(bcff_vars)
print(cond_level_table[:4, :])
```

### 2. Conditional yield change (data)

```python
cond_change_table = np.zeros((8, numb_bcff_vars))

for ii, end, start in zip(range(numb_bcff_vars), bcff_vars_end, bcff_vars_start):
    nw_se = NW_se(data_survey[end] - data_survey[start], 4)
    cond_change_table[:4, ii] = nw_se[0][0], nw_se[0][1], nw_se[1][0], nw_se[1][1]

print('Conditional change (table)')
print(bcff_vars_end)
print(cond_change_table[:4, :])
```

### 3. Conditional change in spread (data)

```python
cond_spread_change_table = np.zeros((8, 2))

for ii, end, start in zip(range(2), bcff_vars_end[1:], bcff_vars_start[1:]):
    spread = data_survey[end] - data_survey[bcff_vars_end[0]] - data_survey[start] + data_survey[bcff_vars_start[0]]
    nw_se = NW_se(spread, 4)
    cond_spread_change_table[:4, ii] = nw_se[0][0], nw_se[0][1], nw_se[1][0], nw_se[1][1]

print('Conditional spread change (table)')
print('medium', 'long')
print(cond_spread_change_table[:4, :])
```

### 4. Subjective forecast error (data)

First we construct a new dataset called `forecast_error` that contains the following things
* survey expectations using the usual naming convention
* realizations of yields shifted $h$ period back so that in period $t$ the variables contains its $t+h$ value. The variable name for a 5 year maturity yield in t plus 4 period is `5.0y_tp4`
* subjective forecast error computed from the above two objects. I identify these variables by appending an `_err` to their usual varnames

Then I merge `forecast_error` with the `data_composite` dataframe

```python
forecast_error = pd.DataFrame(index=data_survey.index)

for i, ytm in enumerate([str(i) + 'y'  for i in np.arange(1, 61)/2]):
    for t, hor in enumerate(horizons):
        d = np.vstack([data_survey[ytm + '_LD'].iloc[(t+1):],
                       data_survey[ytm + hor].iloc[:-(t+1)].values - 
                       data_survey[ytm + '_LD'].iloc[(t+1):].values]).T
        dd = pd.DataFrame(data=d,
                      columns=[ytm + '_tp' + str(t+1), ytm + hor + '_err'],
                      index=date_index[:-(t+1)])
        forecast_error = pd.concat([forecast_error, dd], axis=1)
        
data_survey = pd.concat([data_survey, forecast_error], axis=1)
```

```python
subj_forecast_error_mean = np.zeros((2*numb_yields, 2*numb_hors))
subj_forecast_error_stdev = np.zeros((2*numb_yields, 2*numb_hors))

bias_var = np.zeros((numb_yields, numb_hors))
bias_se =  np.zeros((numb_yields, numb_hors))
bias_var_se =  np.zeros((numb_yields, numb_hors))

for i, ytm in enumerate(list_of_yields):
    for t, hor in enumerate(horizons):
        store = NW_se(data_survey[ytm + hor + '_err'], 4)
        
        subj_forecast_error_mean[2*i, t] = store[0][0]          # mean estimate
        subj_forecast_error_mean[2*i + 1, t] = store[1][0]      # mean standard error
        subj_forecast_error_stdev[2*i, t] = store[0][1]         # stdev estimate
        subj_forecast_error_stdev[2*i + 1, t] = store[1][1]     # stdev standard error

print('Mean subjective forecast error')
print(horizons)
print(subj_forecast_error_mean[:, :numb_hors])
print('')
print('Stdev of subjective forecast error')
print(horizons)
print(subj_forecast_error_stdev[:, :numb_hors])
```

## Corresponding Model implied moments

```python
# Make sure that we use the same zero-coupon yields as in the above analysis
ytm_in_quarter = []

for i, var in enumerate(bcff_vars):
    maturity_years = int(var.rsplit('.', 1)[0])      # pull out the ytm from the var_name
    ytm_in_quarter.append(4*maturity_years)          # turn it into quarters
    
# For the unconditional moments, we evaluate these with the baseline distribution
Ex, Exx = np.zeros((N + 1, 1)), np.zeros((N + 1, N + 1))
Ex[0, 0] = 1
Exx[1:, 1:] = model.from_cont_to_discrete(worstcase = 0)[6]
```

### 1. Conditional level (model)

$E_t y^{(ytm)}_{t + \tau}$ is an affine function of $X_t$

```python
# Coefficients of the affine function 
#   rows:   assets 
#   cols:   first is intercept, the others are for the slope
abb = model.yield_forecast_ab(ytm_in_quarter, tau_end = 4, worstcase = 1)
abb_RE = model.yield_forecast_ab(ytm_in_quarter, tau_end = 4, worstcase = 0, worstcase_abb = 0)

# Model implied subjective expectation
cond_level_table[4, :] = (abb @ Ex).squeeze()
cond_level_table[5, :] = np.sqrt(np.diag(abb @ Exx @ abb.T))

# Rational expectations
cond_level_table[6, :] = (abb_RE @ Ex).squeeze()
cond_level_table[7, :] = np.sqrt(np.diag(abb_RE @ Exx @ abb_RE.T))

print('Estimated robust model')
print(cond_level_table[4:6, :])
print('')
print('Model with rational expectations')
print(cond_level_table[6:8, :])
```

### 2. Conditional change (model)

$E_t \left[y^{(ytm)}_{t + \tau_{end}} - y^{(ytm)}_{t + \tau_{start}}\right]$ is an affine function of $X_t$

```python
tau_end = 5        # We set the benchmark horizon 1 (i.e. from t+1 onward)

abb = model.yield_forecast_ab(ytm_in_quarter, tau_end=tau_end, tau_start=1, worstcase=1)
abb_RE = model.yield_forecast_ab(ytm_in_quarter, tau_end=tau_end, tau_start=1, worstcase=0)

# Model implied subjective expectation
cond_change_table[4, :] = (abb @ Ex).squeeze()
cond_change_table[5, :] = np.sqrt(np.diag(abb @ Exx @ abb.T))

# Rational expectations
cond_change_table[6, :] = (abb_RE @ Ex).squeeze()
cond_change_table[7, :] = np.sqrt(np.diag(abb_RE @ Exx @ abb_RE.T))

print('Estimated robust model')
print(cond_change_table[4:6,:])
print('')
print('Model with rational expectations')
print(cond_change_table[6:8,:])
```

### 2. Conditional change in spread (model)

$E_t \left[y^{(ytm+5)}_{t + \tau_{end}}-y^{(ytm)}_{t + \tau_{end}} - y^{(ytm+5)}_{t + \tau_{start}}+y^{(ytm)}_{t + \tau_{start}}\right]$ is an affine function of $X_t$

```python
#abb = model.yield_forecast_ab(ytm_in_quarter, tau_end=tau_end, tau_start=1, worstcase=1)
#abb_RE = model.yield_forecast_ab(ytm_in_quarter, tau_end=tau_end, tau_start=1, worstcase=0)

# Model implied subjective expectation
cond_spread_change_table[4, 0] = ((abb[1, :]-abb[0, :]) @ Ex).squeeze()
cond_spread_change_table[4, 1] = ((abb[2, :]-abb[0, :]) @ Ex).squeeze()
cond_spread_change_table[5, 0] = np.sqrt((abb[1, :]-abb[0, :]) @ Exx @ (abb[1, :]-abb[0, :]).T)
cond_spread_change_table[5, 1] = np.sqrt((abb[2, :]-abb[0, :]) @ Exx @ (abb[2, :]-abb[0, :]).T)

# Rational expectations
cond_spread_change_table[6, 0] = ((abb_RE[1, :]-abb_RE[0, :]) @ Ex).squeeze()
cond_spread_change_table[6, 1] = ((abb_RE[2, :]-abb_RE[0, :]) @ Ex).squeeze()
cond_spread_change_table[7, 0] = np.sqrt((abb_RE[1, :]-abb_RE[0, :]) @ Exx @ (abb_RE[1, :]-abb_RE[0, :]).T)
cond_spread_change_table[7, 1] = np.sqrt((abb_RE[2, :]-abb_RE[0, :]) @ Exx @ (abb_RE[2, :]-abb_RE[0, :]).T)

print('Estimated robust model')
print(cond_spread_change_table[4:6,:])
print('')
print('Model with rational expectations')
print(cond_spread_change_table[6:8,:])
```

### 4. Subjective forecast error (model)

Subjective forecast error = bias + objective forecast error

I am working with the discrete time matrices:
 * Value of state in period $t+n$
 $$X_{t+n} = \sum_{i=0}^{n-1} \left(\kappa^D\right)^{i} \phi^D + \left(\kappa^D\right)^{n}X_t + \sum_{j=1}^{n} \left(\kappa^D\right)^{n-j}\sigma\varepsilon_{t+j}$$
 so 
  $$E[X_{t+n}|X_t] = \sum_{i=0}^{n-1} \left(\kappa^D\right)^{i} \phi^D + \left(\kappa^D\right)^{n}X_t\quad\quad\text{and}\quad\quad Var[X_{t+n}| X_t] = \sum_{j=1}^{n} \left(\kappa^D\right)^{n-j}\sigma\sigma^T\left[\left(\kappa^D\right)^{n-j}\right]^T$$
  and 
  $$\tilde{E}[X_{t+n}|X_t] = \sum_{i=0}^{n-1} \left(\tilde{\kappa}^D\right)^{i} \tilde{\phi}^D + \left(\tilde{\kappa}^D\right)^{n}X_t\quad\quad\text{and}\quad\quad \widetilde{Var}[X_{t+n}| X_t] = \sum_{j=1}^{n} \left(\tilde{\kappa}^D\right)^{n-j}\sigma\sigma^T\left[\left(\tilde{\kappa}^D\right)^{n-j}\right]^T$$

 * **Bias**: 
 $$\text{Bias}(X_t) := \tilde{E}[X_{t+n}|X_t] - E[X_{t+n}|X_t]$$
 
 * **Subjective forecast error:** 
 $$ SFE := \text{Bias}(X_t) - \sum_{j=1}^{n} \left(\kappa^D\right)^{n-j}\sigma\varepsilon_{t+j} $$
 where the two terms are independent given that $X_t$ is independent of $\varepsilon_{t+j}$ for all $j\geq 1$
 
$\Rightarrow\quad\quad$ Var(SFE) = Var(Bias) + Var(Objective Forecast error)

$\Rightarrow\quad\quad$ ratio of bias variance = Var(Bias)/(Var(Bias) + Var(Objective Forecast error))


```python
# Bias term (unconditional under the baseline)
bias_mean_model = np.zeros((numb_yields, 5))
bias_stdev_model = np.zeros((numb_yields, 5))

# Objective forecast error (unconditional under the true -> mean is zero)
obj_forecast_error_stdev_model = np.zeros((numb_yields, 5))

# Ratio of bias variance
ratio_bias_variance = np.zeros((numb_yields, 5))

bb = model.zero_coupon_yields(ytm_in_quarter, worstcase = 1)[1:, :].T  # assets are in rows

for i, tau in enumerate(range(1, 6)):
    abb = model.yield_forecast_ab(ytm_in_quarter, tau_end=tau, worstcase = 1)
    abb_obj = model.yield_forecast_ab(ytm_in_quarter, tau_end=tau, worstcase = 0)
    abb_bias = abb - abb_obj
    
    # Model implied bias
    bias_mean_model[:, i] = (abb_bias @ Ex).squeeze()
    bias_stdev_model[:, i] = np.sqrt(np.diag(abb_bias @ Exx @ abb_bias.T))

    # Model implied objective forecast error (std)
    obj_forecast_error_stdev_model[:, i] = np.sqrt(np.diag(bb @ model.X_CondDist(Ex[1:, 0], tau).cov @ bb.T))
    
    # Ratio of bias variance
    ratio_bias_variance[:, i] = bias_stdev_model[:, i]**2/(bias_stdev_model[:, i]**2 + 
                                                             obj_forecast_error_stdev_model[:, i]**2)
    
# Create Table 4 from the paper
for i, ytm in enumerate(list_of_yields):
    for t, hor in enumerate(horizons):
        subj_forecast_error_mean[2*i, numb_hors + t] = bias_mean_model[i, t]
        subj_forecast_error_stdev[2*i, numb_hors + t] = obj_forecast_error_stdev_model[i, t]
```

## Predictive regression to approximate the "objective" model of yields


Project vector $Y_{t+h}$ onto $Y_t$ with OLS where $Y=[1y, 2y, 3y, 4y, 5y, 10y, 15y]$

```python
#varlist = ['TRNPY01', 'TRNPY03', 'TRNPY05', 'TRNPY10']
#ytm_years = [1, 3, 5, 10]
#X = data[varlist]['1962':]

varlist = ['SVENY01', 'SVENY02', 'SVENY03', 'SVENY04', 'SVENY05', 'SVENY07', 'SVENY10', 'SVENY15']
ytm_years = [1, 2, 3, 4, 5, 7, 10, 15]
X = data[varlist]

forecast_error = pd.DataFrame(index=X.index)

for t in range(5):
    d = X.iloc[(t+1):].values
    dd = pd.DataFrame(data=d, columns=[ytm + '_tp' + str(t+1) for ytm in varlist],
                          index=X.index[:-(t+1)])
    forecast_error = pd.concat([forecast_error, dd], axis=1)
        

obj_forecast = pd.DataFrame(index=X.index)
for var in varlist:
    obj_forecast_hor = pd.DataFrame(index=X.index)

    for hor in range(1, 6):
        Y = forecast_error[var + '_tp' + str(hor)]
        X = sm.add_constant(X)
        forcasting_regression = sm.OLS(Y, X, missing='drop')
        res = forcasting_regression.fit()
        dd = pd.concat([res.fittedvalues, -res.resid], axis=1)
        dd.columns = [var + '_' + str(hor) + 'q_obj', var + '_' + str(hor) + 'q_err_obj']
        obj_forecast_hor = pd.concat([obj_forecast_hor, dd], axis=1)
        
    obj_forecast = pd.concat([obj_forecast, obj_forecast_hor], axis=1)
    
forecast_error = pd.concat([X, forecast_error, obj_forecast], axis=1)
```

```python
data_bias = pd.DataFrame(index=forecast_error['1982-12':].index)

for ytm, var in zip(ytm_years, varlist):
    for h in range(1, 6):
        if ytm < 10:
            vv = var[:-1]
        else:
            vv = var[:-2]
        bs = data_survey[str(ytm)+'.0y_'+str(h)+'q']['1982-12':]-\
                        forecast_error[vv+str(ytm)+'_'+str(h)+'q_obj']['1982-12':]
        bm = (data_model[str(ytm)+'.0y_'+str(h)+'q']-data_model[str(ytm)+'.0y_'+str(h)+'q_obj'])['1982-12':]
        ff = pd.concat([bs, bm], axis=1)
        ff.columns = [str(ytm) + '.0y_' + str(h) + 'q_bias_survey', str(ytm) + '.0y_' + str(h) + 'q_bias_model']
        ff.index = forecast_error['1982-12':].index
        
        data_bias = pd.concat([data_bias, ff], axis=1)
```

## Replication Piazzesi et al (2015)

Roughly speaking, they find that the dynamics of $[y^{(1)}_t, \Delta y^{(5)}_t]$ estimated from data is less persistent than what's needed to replicate survey expectations.

They use a linear model for $xx = [y^{(1)}_t, \Delta y^{(5)}_t]$ so that the "objective" forecast is

$$E_t y^{(1)}_{t+n} = Ey^{(1)} + A^{n}xx_t $$

Then they look for a $\gamma_0$ and $\gamma_1$ such that  

$$\tilde{E}_t y^{(1)}_{t+n} = (Ey^{(1)} + \gamma_0) + (A + \gamma_1)^{n} xx_t $$

where the tilde is for the survey expectations.


#### First attempt: model implied autocorrelation function for $y^{1}_t$ and $y^{20}_t$

#### Second attempt: VAR(1) for yield and spread

```python
def yield_corr(Ex, Exx, p, k, b, db):
    """
    demeaned yield series
    """
    
    um = b.T @ Ex
    ur = db.T @ Ex
    mm = b.T @ Exx @ b
    mr = db.T @ Exx @ b
    bb = db.T @ Exx @ db
    
    Eytyt = np.asarray([[1, um, ur],
                        [um, mm, mr],
                        [ur, mr, bb]])
    
    a_bar, da_bar = b.T @ p, db.T @ p
    b_bar, db_bar = k.T @ b, k.T @ db
    
    UL = a_bar + b_bar.T @ Ex
    UR = da_bar + db_bar.T @ Ex
    ML = a_bar*b.T @ Ex + b.T @ Exx @ b_bar
    MR = da_bar*b.T @ Ex + b.T @ Exx @ db_bar
    BL = a_bar*db.T @ Ex + db.T @ Exx @ b_bar
    BR = da_bar*db.T @ Ex + db.T @ Exx @ db_bar
    
    Eytyt1 = np.asarray([[UL[0,0], UR[0,0]],
                         [ML[0,0], MR[0,0]],
                         [BL[0,0], BR[0,0]]])

    
    return (np.linalg.inv(Eytyt) @ Eytyt1).T
```

```python
def obj_func(params, results, subj_exp, xx):
    
    eta0 = params[:2].reshape(1, 2)
    eta1 = params[2:].reshape(2, 2)

    mu = results.mean() + eta0
    A = results.params[1:].values.T + eta1
    
    obj_forecast = (np.linalg.matrix_power(A, 5) @ (xx - mu).T).T + mu
    
    return np.nanmean((obj_forecast - subj_exp.values)**2), mu, A, obj_forecast
```
